```python
# Q 11: Identify and remove outliers in the km_driven column using the IQR method.

# IQR method for outlier detection and removal
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]
print("Outliers identified:")
print(outliers['km_driven'])

# Remove outliers
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]
```

```python
# Q 12: Perform a bivariate analysis to visualize the relationship between year and selling_price.

# Scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='year', y='selling_price')
plt.title('Selling Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()

# Boxplot
plt.figure(figsize=(15, 7))  # Adjust figure size for better readability
sns.boxplot(data=df, x='year', y='selling_price')
plt.title('Selling Price Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout() #prevents x axis labels from being cut off
plt.show()

# Line plot (aggregate: mean selling price by year)
yearly_mean_price = df.groupby('year')['selling_price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_mean_price, x='year', y='selling_price')
plt.title('Mean Selling Price Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Selling Price')
plt.show()
```

```python
# Q 13: What is the average depreciation in selling price based on the bike's age (current year - manufacturing
# year).

import datetime

# Get the current year
current_year = datetime.datetime.now().year

# Calculate the bike's age
df['age'] = current_year - df['year']

# Calculate average selling price for each age
average_price_by_age = df.groupby('age')['selling_price'].mean().reset_index()

# Calculate depreciation (difference in average price from the previous year)
average_price_by_age['depreciation'] = average_price_by_age['selling_price'].diff()

# Display the results
print(average_price_by_age)

# Visualize the average selling price by age
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_price_by_age, x='age', y='selling_price', marker='o')
plt.title('Average Selling Price vs. Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Average Selling Price')
plt.grid(True)
plt.show()

# Visualize the depreciation
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_price_by_age, x='age', y='depreciation', marker='o')
plt.title('Depreciation vs. Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Depreciation (Average Selling Price Change)')
plt.grid(True)
plt.show()

# Calculate overall average depreciation
overall_average_depreciation = average_price_by_age['depreciation'].mean()
print(f"\nOverall Average Depreciation: {overall_average_depreciation}")

# Calculate average percentage depreciation
average_price_by_age['percentage_depreciation'] = (average_price_by_age['depreciation'] / average_price_by_age['selling_price'].shift(1)) * 100
overall_average_percentage_depreciation = average_price_by_age['percentage_depreciation'].mean()
print(f"Overall Average Percentage Depreciation: {overall_average_percentage_depreciation}%")
```

```python
# Q 14: Which bike names are priced significantly above the average price for their manufacturing year?

average_price_by_year = df.groupby('year')['selling_price'].mean().reset_index()
average_price_by_year.rename(columns={'selling_price': 'avg_selling_price_year'}, inplace=True)

# Merge the average price with the original DataFrame
df = pd.merge(df, average_price_by_year, on='year')

# Calculate the difference between the selling price and the average price for the year
df['price_diff'] = df['selling_price'] - df['avg_selling_price_year']

# Identify bikes with significantly higher prices (e.g., above a threshold)
threshold_multiplier = 2 #Adjust this to change how many standard deviations above the mean to filter.
threshold = df['avg_selling_price_year'] + (df['selling_price'].std() * threshold_multiplier)

high_priced_bikes = df[df['selling_price'] > threshold]

# Display the results
if not high_priced_bikes.empty:
    print("Bikes priced significantly above the average for their manufacturing year:")
    print(high_priced_bikes[['name', 'year', 'selling_price', 'avg_selling_price_year', 'price_diff']])
else:
    print("No bikes found with prices significantly above the average for their manufacturing year.")
```

```python
# Q 15: Develop a correlation matrix for numeric columns and visualize it using a heatmap.

numeric_columns = df.select_dtypes(include=['number'])

if numeric_columns.empty:
    print("No numeric columns found in the dataset.")
    exit()

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```

```python
# Q 1: What is the average selling price of cars for each dealer, and how does it compare across different dealers?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/Car Sale.csv')

# Filter for dealer sales
dealer_sales = df[df['Dealer_Name'] == 'Dealer']

if dealer_sales.empty:
    print("No dealer sales found in the dataset.")
    exit()

# Calculate the average selling price for each dealer
average_price_by_dealer = dealer_sales.groupby('Dealer_Name')['Price ($)'].mean()

# Display the average selling price
print("Average Selling Price for Dealers:")
print(average_price_by_dealer)

# Visualize the average selling price
plt.figure(figsize=(8, 6))
sns.barplot(x=average_price_by_dealer.index, y=average_price_by_dealer.values)
plt.title('Average Selling Price for Dealers')
plt.xlabel('Dealer Type')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()
plt.show()

# If you have a column that identifies individual dealers (e.g., dealer_name), you can do a more granular analysis:
if 'dealer_name' in dealer_sales.columns:
    average_price_by_individual_dealer = dealer_sales.groupby('dealer_name')['selling_price'].mean().sort_values(ascending=False)
    print("\nAverage Selling Price for Individual Dealers:")
    print(average_price_by_individual_dealer)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=average_price_by_individual_dealer.index, y=average_price_by_individual_dealer.values)
    plt.title('Average Selling Price for Individual Dealers')
    plt.xlabel('Dealer Name')
    plt.ylabel('Average Selling Price')
    plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate x-axis labels, adjust font size
    plt.tight_layout()
    plt.show()

else:
    print("\n'dealer_name' column not found, cannot perform individual dealer analysis.")
```

```python
# Q 2: Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
# trendsE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/Car Sale.csv')

# Calculate price variation (standard deviation) for each company
price_variation = df.groupby('Model')['Price ($)'].std().sort_values(ascending=False)

# Display the company with the highest price variation
highest_variation_company = price_variation.index[0]
highest_variation_std = price_variation.iloc[0]

print(f"The car brand with the highest price variation is '{highest_variation_company}' with a standard deviation of {highest_variation_std:.2f}.")

# Visualize price variation
plt.figure(figsize=(12, 6))
sns.barplot(x=price_variation.index, y=price_variation.values)
plt.title('Price Variation (Standard Deviation) by Car Brand')
plt.xlabel('Car Brand (Company)')
plt.ylabel('Standard Deviation of Selling Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize the price distribution for the highest variation company
highest_variation_df = df[df['Model'] == highest_variation_company]

plt.figure(figsize=(10, 6))
sns.histplot(highest_variation_df['Price ($)'], kde=True)
plt.title(f'Price Distribution for {highest_variation_company}')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q 3:  What is the distribution of discounted prices vs. actual prices?

def plot_price_distributions(df, actual_price, discounted_price):
    """
    Plots the distributions of actual prices and discounted prices.

    Args:
        df (pd.DataFrame): The DataFrame containing price data.
        actual_price_col (str): The name of the column containing actual prices.
        discounted_price_col (str): The name of the column containing discounted prices.
    """

    plt.figure(figsize=(12, 6))

    # Plot distribution of actual prices
    plt.subplot(1, 2, 1)
    sns.histplot(df[actual_price], kde=True)
    plt.title('Distribution of Actual Prices')
    plt.xlabel('Actual Price')
    plt.ylabel('Frequency')

    # Plot distribution of discounted prices
    plt.subplot(1, 2, 2)
    sns.histplot(df[discounted_price], kde=True)
    plt.title('Distribution of Discounted Prices')
    plt.xlabel('Discounted Price')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
